```python
# cat va cont ustunlarni ajratib olish:

# category columns
cat_name = []
# raqamli columns
cont_name = []

for col in df.columns:
  if df[col].dtype== object:
    cat_name.append(col)
  else :
    cont_name.append(col)


print(len(cont_name))
print(len(cat_name))
```

`cont_name` listdan `"Status"` ustunini olib tashlaymiz. chunki u bizga bashorat uchun kerak.

```python
del cont_name[-2]
```

```python
cont_name
```

```python
from fastai.tabular.all import *

# path
path = Path('.')

# path.ls()

# dataloaders
dls = TabularDataLoaders.from_csv(path/'Loan_Default.csv', path=path, bs=64, y_names = 'Status',
                                  cat_names = cat_name,
                                  cont_names = cont_name,
                                  procs = [FillMissing , Categorify, Normalize])


# train
learn = tabular_learner(dls, metrics=accuracy)

# tabular data uchun fit_one_cycle(), modelni o dan quradi
learn.fit_one_cycle(3)
```

```python
learn.show_results(max_n=64)
```
